```python
plt.subplot(2,4,1)
sns.distplot(data['Temp_C'], color = 'red')
plt.xlabel('Ratio of temperature', fontsize = 12)
plt.grid()

plt.subplot(2,4,2)
sns.distplot(data['Dew Point Temp_C'], color = 'yellow')
plt.xlabel('Ratio of dew point', fontsize = 12)
plt.grid()

plt.subplot(2,4,3)
sns.distplot(data['Rel Hum_%'], color = 'lightblue')
plt.xlabel('Ratio of humidity', fontsize = 12)
plt.grid()

plt.subplot(2,4,4)
sns.distplot(data['Wind Speed_km/h'], color = 'lightgreen')
plt.xlabel('Ratio of wind speed', fontsize = 12)
plt.grid()

plt.subplot(2,4,5)
sns.distplot(data['Visibility_km'], color = 'pink')
plt.xlabel('Ratio of visibility', fontsize = 12)
plt.grid()

```

```python
print("Fog/snow")
print(data[(data['Temp_C'] < 0) & (data['Rel Hum_%'] > 50)])
print("---------------------------------------------------------------------------")
print("Thunderstorms/ Rain")
print(data[(data['Temp_C'] < -4) & (data['Rel Hum_%'] > 80)])
print("---------------------------------------------------------------------------")
print("Normal weather")
print(data[(data['Weather'] == 'Mainly Clear') & (data['Rel Hum_%'] < 30)])
print("---------------------------------------------------------------------------")
```
